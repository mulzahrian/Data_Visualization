```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import pandas as pd

data = pd.read_csv('/content/sample_data/bike_share.csv')

```

```python
data.columns
```

```python
plt.hist(data['temp'], bins=20, color='blue')
plt.xlabel('Suhu (temp)')
plt.ylabel('Frekuensi')
plt.title('Histogram Suhu')
plt.show()
```

```python
plt.scatter(data['temp'], data['windspeed'], color='green')
plt.xlabel('Suhu (temp)')
plt.ylabel('Kecepatan Angin (windspeed)')
plt.title('Scatter Plot antara Suhu dan Kecepatan Angin')
plt.show()
```

```python
season_counts = data['season'].value_counts()
season_names = ['Spring', 'Summer', 'Fall', 'Winter']
plt.bar(season_names, season_counts, color='orange')
plt.xlabel('Musim (season)')
plt.ylabel('Jumlah Data')
plt.title('Bar Plot untuk Musim')
plt.show()
```

```python
data.boxplot(column='total_rentals', by='weekday', showfliers=False)
plt.xlabel('Hari dalam Seminggu')
plt.ylabel('Total Sewaan')
plt.title('Box Plot Total Sewaan per Hari dalam Seminggu')
plt.suptitle('')  # Menghilangkan judul atas yang dibuat otomatis
plt.xticks(range(1, 8), ['Sun', 'Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat'])  # Mengganti label hari
plt.show()
```
